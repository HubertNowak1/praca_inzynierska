```python
df7 = pd.read_csv(
    "data_wid.csv",
    sep=";",
    engine="python",
    quotechar='"',
    skiprows=1,
    header=None
)
```

```python
df7.columns = df7.iloc[0] 
df7 = df7.iloc[1:].reset_index(drop=True)

df7 = df7.melt(
    id_vars=["Percentile ", "Year "],
    var_name="raw_column",
    value_name="value"
)

df7["country"] = (
    df7["raw_column"]
    .str.split("\n")
    .str[-1]
    .str.strip()
)

df7["indicator_code"] = (
    df7["raw_column"]
    .str.extract(r"^([a-z0-9_]+)_")
)

indicator_map = {
    "agdpro_992_i": "gdp_pc",
    "sptinc_z": "income_share"
}


df7 = df7[df7["indicator_code"].isin(indicator_map)]
df7["indicator"] = df7["indicator_code"].map(indicator_map)

df7["Percentile"] = df7["Percentile "].str.strip()
df7["value"] = pd.to_numeric(df7["value"], errors="coerce")
df7["Year"] = df7["Year "].astype(int)
```

```python
df_income = (
    df7
    .pivot_table(
        index=["country", "Year"],
        columns="Percentile",
        values="value"
    )
    .reset_index()
)

df_income = df_income.drop(columns=["pall"])

gdp = (
    df7[df7["indicator"] == "gdp_pc"]
    [["country", "Year", "value"]]
    .rename(columns={"value": "gdp_pc"})
)

df7 = df_income.merge(
    gdp,
    on=["country", "Year"],
    how="left"
)

df7 = (
    df7
    .drop_duplicates(subset=["country", "Year"], keep="first")
    .reset_index(drop=True)
)
```

```python
df7 = df7.rename(columns={
    "country": "Kraj",
    "Year": "Rok",
    "gdp_pc": "PKB_pc"
})
```
